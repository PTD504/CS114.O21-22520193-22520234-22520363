```python
# import các thư viện
import numpy as np
import pandas as pd
import keras
from keras import layers
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from keras.applications.vgg16 import VGG16, preprocess_input
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import img_to_array, load_img, ImageDataGenerator
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import os, csv, cv2, random as rd
from PIL import Image

import warnings
warnings.filterwarnings("ignore")
```

```python
# Các biến toàn cục
Category_map = {
  'Others': 0,
  'Honda': 1,
  'Suzuki': 2,
  'Yamaha': 3,
  'VinFast': 4
}

# thư mục gốc chứa thư mục dữ liệu
# szWorkingDir = 'drive/MyDrive/Public'
# Đường dẫn đến các file csv chưa làm sạch
szWorkingDir = '/kaggle/working/Public'
szInputDir = '/kaggle/input/traintest-data'

# dùng để hiển thị
nNumImgsPerRow = 10
nImgHeight = nImgWidth = 150
```

```python
# Hàm kiểm tra tính hợp lệ của ảnh
def is_valid_image(file_path):
    try:
        _ = Image.open(file_path)
        return True
    except:
        return False

# Hàm để đọc file CSV và loại bỏ các ảnh không hợp lệ
def load_and_filter_data_from_csv(csv_file, base_dir):
    valid_file_paths = []
    valid_labels = []

    df = pd.read_csv(os.path.join(base_dir, csv_file))
    for index, row in df.iterrows():
        file_path = os.path.join(szWorkingDir, row[0])
        if is_valid_image(file_path):
            valid_file_paths.append(file_path)
            valid_labels.append(row[1])

    return valid_file_paths, valid_labels
```

### Preprocessing Data

```python
def load_and_preprocess_image(image_path, target_size=(224, 224)):
    try:
        img = load_img(image_path, target_size=target_size)
        img_array = img_to_array(img)
        img_array = preprocess_input(img_array)
        return img_array
    except Exception as e:
        return None
```

### Data Augmentation

```python
# Khởi tạo ImageDataGenerator với các phép biến đổi tăng cường dữ liệu
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

### Load Data

```python
def batch_generator(file_paths, labels, batch_size, target_size=(224, 224), datagen=None):
    num_samples = len(file_paths)
    num_batches = num_samples // batch_size  # Số lượng batch cần tạo

    for i in range(num_batches):
        batch_paths = file_paths[i * batch_size: (i + 1) * batch_size]
        batch_labels = labels[i * batch_size: (i + 1) * batch_size]

        images = [load_and_preprocess_image(path, target_size) for path in batch_paths]
        images = [img for img in images if img is not None]

        if len(images) != batch_size:
            continue

        images = np.array(images)

        if datagen:
            augmented_images = []
            for img in images:
                augmented_images.append(datagen.random_transform(img))
            images = np.array(augmented_images)

        yield images, np.array(batch_labels)

    if num_samples % batch_size != 0:
        batch_paths = file_paths[num_batches * batch_size:]
        batch_labels = labels[num_batches * batch_size:]

        images = [load_and_preprocess_image(path, target_size) for path in batch_paths]
        images = [img for img in images if img is not None]

        if images:
            images = np.array(images)

            if datagen:
                augmented_images = []
                for img in images:
                    augmented_images.append(datagen.random_transform(img))
                images = np.array(augmented_images)

            yield images, np.array(batch_labels)
```

### Feature Extraction

```python
# Trích xuất đặc trưng từ mô hình pre-trained VGG16
def extract_features(model, generator, steps):
    features = []
    for i in range(steps):
        images, _ = next(generator)
        features_batch = model.predict(images)
        features.append(features_batch)
    return np.vstack(features)
```

```python
# Mô hình Logistic Regression với hầu hết các tham số mặc định
log_reg = LogisticRegression(max_iter=1000)
```
